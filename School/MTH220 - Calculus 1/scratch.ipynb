```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the main graph
x_main = np.array([1, 2, 3, 4, 5])
y_main = np.array([10, 8, 12, 6, 9])

# Sample data for the bars under the graph
x_bars = np.array([1, 2, 3, 4, 5])
y_bars = np.array([2, 3, 1, 4, 2]) # Heights of the bars

# Create the figure and axes
fig, ax = plt.subplots()

# Plot the main graph (e.g., a line plot)
ax.plot(x_main, y_main, marker='o', linestyle='-', color='blue', label='Main Data')

# Add bars under the graph
# The 'bottom' parameter is set to a negative value to place the bars below the x-axis
# The height of the bars will extend from this negative 'bottom' value upwards by 'y_bars'
ax.bar(x_bars, y_bars, bottom=-5, color='red', alpha=0.7, label='Bars Under Graph')

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Graph with Bars Underneath')
ax.legend()
ax.grid(True)

# Show the plot
plt.show()
```
