$$
\Large
\int_a^b f(x)\,dx \, \approx \quad
\frac{\textcolor{magenta}{\Delta x}}{3} 

\left[
f(x_0)
+ \sum_{i=1}^{n-1} [ \textcolor{cyan}{k_i} \cdot f(x_i) ]
+ f(x_n)
\right]

\quad \left|

\begin{aligned}
&\textcolor{magenta}{\Delta x} = \frac{b-a}{n} \\[0.3em]
&\textcolor{cyan}{k_i} =
\begin{cases}
4, & \text{if } i \text{ is odd} \\
2, & \text{if } i \text{ is even}
\end{cases}
\end{aligned}

\right.
$$

```python
expr = 9+x**3
f = sp.lambdify(x, expr)
n = 2
a, b = 0, 4

points = np.linspace(a,b, n+1) # n intervals, so n+1 points
delta = (b-a)/n
evaluations = []
for i, point in enumerate(points):
    if i == 0 or i == n:
        evaluations.append(f(point))
    else:
        k = 2 + 2*(i%2) # 4 if odd, 2 if even
        evaluations.append(k*f(point))

area = delta/3*sum(evaluations)
area
```

```python
import sympy as sp
x = sp.symbols('x')

f = x
g = sp.root(x+6, 4)

# Compute the integral directly
integral_direct = sp.integrate(f * g, x)
integral_direct.together().factor()
```

```python

# Integration by parts manually:
# Let u = log(x), dv = x^2 dx
u = f
du = sp.diff(u, x)
dv = g
v = sp.integrate(dv, x)

# Apply integration by parts: u dv = uv - v du
parts_result = u * v - sp.integrate(v * du, x)

parts_result
```
