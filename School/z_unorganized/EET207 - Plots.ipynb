```python
import matplotlib.pyplot as plt
```

### Question 1: Plot the following on one graph:
x=[1,10,100]  
y=[2,20,200]  
  
a.	Turn on the grid for the graph.  
b.	Make  the title “Line Plot for X and Y”.  
c.	Make the x-label ‘x-axis’ and the y-label ‘y-axis’.  
d.	Mark the color of the line on the graph a red.  
e.	Create a legend that shows the label VI characteristics.  


```python
x=[1,10,100]  
y=[2,20,200] 

# Create figure and axes objects
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y, color='red', label='VI characteristics')

# Set title and labels
ax.set_title('Line Plot for X and Y')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

# Enable grid
ax.grid(True)

# Add legend
ax.legend()

# Show the plot
plt.show()
```

## Question 2: Plot the following two subplots with 4 rows and 2 colums:
`Subplot1 (location 1)`  
x=[1,10,100]  
y1=[2,20,200]  
  
`Subplot2 (location 6)`  
x=[1,10,100]  
y2=[3,30,300]  
  
a.	Turn on the grid for the graph.  
b.	Make  the title “Line Plot for X and Y”.  
c.	Make the x-label ‘x-axis’ and the y-label ‘y-axis’.  
d.	Mark the color of the line on the graph a red and second line as blue.  
e.	Create a legend that shows the label VI characteristics and second I-V  


```python
x=[1,10,100]  
y1=[2,20,200] 
y2 = [3,30,300]

fig, axes = plt.subplots(4,2)


# Plot the data
axes[0,0].plot(x, y1, color='red', label='VI characteristics')
axes[3,1].plot(x, y2, color='blue', label='second I-V')

# Set title and labels
my_axes = [axes[0,0],axes[3,1]]
for ax in my_axes:
    ax.set_title('Line Plot for X and Y')
    ax.set_xlabel('x-axis')
    ax.set_ylabel('y-axis')
    ax.grid(True) # Enable grid


# Add legend
fig.legend() #On figure to make things neater

# Turn off empty axes
for ax in axes.flat:
    if not ax.has_data():
        ax.axis('off')

# Show the plot
plt.show()
```
