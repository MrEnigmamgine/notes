```python
#Example
W = 10 #ton
alpha = 40 #
v1 = 80 #mph
```

Formulas:

$ E_K = \frac{1}{2} m v^2 $  
$ E_P = mgh $  
$ E_M = E_K + E_P $  
$ \Delta E_M + L = 0 $  
$ m = \frac{W}{g} $

```python
#Given
W = 13 #ton
alpha = 58 #
v1 = 97 #mph

#Assumed
g = 32.2 #m/s/s
E_p1 = 0
E_k2 = 0

#Unit Conversions
k_mph_fps = 5280/3600
k_ton_lb = 2000

#Unknown
d = sp.symbols('d')
```

```python
v1 =    v1 *k_mph_fps
W  =    W  *k_ton_lb
```

```python
m = W/g
m #slugs
```

Acceleration method

$ v^2 = v_i^2 + 2as \xrightarrow{\text{Solve for } s} s = \frac{\vec{v_f}^{2} - \vec{v_i}^{2}}{2 \vec{a}} $

```python
theta_g = complement_alpha = 90-alpha
a = g * math.sin(math.radians(alpha))
a = g * math.cos(math.radians(theta_g))
a
```

```python
distance = v1**2/(2*a)
distance
```

Conservation of energy method

```python
h = d*sp.sin(sp.rad(alpha))
h
```

```python
E_k1 = 1/2*m*v1**2
E_k1 #ft-lb
```

```python
E_p2 = m*g*h
E_p2
```

```python
E_m1 = E_k1 + E_p1
E_m2 = E_k2 + E_p2
```

```python
expr = sp.Eq(E_m1, E_m2)
expr
```

```python
sp.solve(expr, d)[0]
```

```python
# Reset Variables
E_k1 = E_k2 = E_p1 = E_p2 = None
```

```python
#Example
k_spring = 340
x = 0.3
```

Formulas:  

$ F = k_sx $  
$ U = Fs $  
$ U_{\text{spring}} = \frac{1}{2}kx^2 $  
$ F = ma $  
$ F_\mu = \mu F_N $  
$ \Delta E_M - U_T = 0 $  

```python
#Given
k_spring = 223 #N/m
k_friction = 0.8 
x = 0.4 #m
slope = 10 #degree
m = 4 #kg

#Assumed
g = 9.81 #m/s/s
KE1 = 0
KE2 = 0
PE2 = 0

#Unknown
d = sp.symbols('d')

#Needed 
PE1 = __
Ut = __
```

Solution:

```python
g_normal = g * math.cos(math.radians(slope)) #Same as sin() of the complement
g_parallel = g * math.sin(math.radians(slope)) #Same as cos() of the complement
F_friction = k_friction * m * g_normal
F_friction # Newtons resisting
```

```python
F_spring = k_spring * x
F_spring #Newtons
```

```python
slope_rads = sp.rad(slope)
h = d*sp.sin(slope_rads)
PE1 = m*g*h
PE1 #Nm
```

```python
ME2 = KE2 + PE2 
ME1 = KE1 + PE1
```

```python
U_spring = 1/2*k_spring*x**2
U_friction = F_friction*d
Ut = U_spring - U_friction
Ut
```

```python
expr = sp.Eq(
    ME1 + Ut,
    ME2
)
expr
```

```python
sp.solve(expr, d)[0]
```

### 15


```python
# Example
Wa = 200 #lb
k_gyration = 2 #ft
h = 8 #ft
```

Formulas:  

$ I = m k_\gamma^2 $  
$ E_K = \frac{1}{2}I\omega^2 $  
$ E_K = \frac{1}{2}mv^2 $  
$ v_T = \omega r $  
$ E_P = mgh $  
$ F = ma $  
$ \Delta E_M - U_T = 0 $  

```python
# Given
Wa = 288 #lb
k_gyration = 2 #ft
h = 7 #ft

# Assumed
g = 32.2 #ft/s/s
r = 3 #ft
Wb = 96.6 #lb
Ut = 0

# Unkown
omega = sp.symbols(r'\omega')
```

Solution

```python
# Find masses of both objects
m_a = Wa/g
m_b = Wb/g
```

```python
# Find moment of inertia of body A
I = m_a * k_gyration**2
I
```

Assume all of state 1's potential energy gets converted to kenetic energy.  
The kenetic energy is split between body A and body B.

```python
PE1 = Wb*h
PE1 #ft-lb
```

```python
KE2a = 1/2 * I * omega**2
KE2a
```

```python
KE2b = 1/2 * m_b * (omega*r)**2 # v = omega * r
KE2b
```

```python
KE2 = KE2a + KE2b
KE2 #ft-lb
```

```python
expr = sp.Eq( PE1, KE2)
sp.solve(expr, omega)[0]
```
