Formula:

$ I_c = m k_\gamma^2 $  
$ \Tau \Delta t = I_c \Delta \omega $  

```python
# Given
W = 122 #lb
k_gyration = 10 #in
rpm_initial = 1620
rpm_final = 733
t = 4 #seconds

# Assumed
g = 32.2 #ft/s/s

# Unit Conversions
k_inch_foot = 1/12
k_rmp_rad = 2*math.pi/60

# Find
Tau = __
```

```python
# Convert units:
k_gyration = k_gyration * k_inch_foot
wi = rpm_initial * k_rmp_rad
wf = rpm_final * k_rmp_rad
```

```python
# Get moment of inertia
m = W/g
I = m * k_gyration**2
```

```python
# Solve for torque
dw = wf-wi
Tau = I*dw/t
Tau
```
