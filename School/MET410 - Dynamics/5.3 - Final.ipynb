Formula:  

$ F_\mu = F_N \mu $  
$ F = ma $  

```python
#Given
m = 58 #kg
mu = 0.27

#Assumed
g = 9.81 #m/s/s
slope = [12,5]

#Unknown
a = __
```

```python
# Convert slope to something we can work with
alpha = math.degrees(math.atan2(*slope[::-1]))
alpha
```

```python
# Accelerating force
theta_g = complement_alpha = 90-alpha
Fp = g*m * math.cos(math.radians(theta_g))
```

```python
# Friction force
Fn = g*m * math.sin(math.radians(theta_g))
Fmu = -Fn*mu 
```

```python
F = Fp + Fmu
a = F/m
highlight(a) #m/s/s
```
