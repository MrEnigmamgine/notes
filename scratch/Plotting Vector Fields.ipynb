```python
import matplotlib.pyplot as plt
import numpy as np
```

# Visualizing Vector Fields
This notebook explores the plotting of vector fields.  
https://www.youtube.com/watch?v=u0hfen9VkJs


$$ \vec{f}(x,y) = \sin(x)\,\hat{x} + \cos(y)\,\hat{y} $$

```python
# Setting up data coordinates
x_range = y_range = np.linspace(0, 2*np.pi, 10) 
x,y = np.meshgrid(x_range,y_range)

# The vector data that corresponds to each data coordinate. 
fx, fy = np.sin(x), np.cos(y)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8,8)
ax.set_aspect('equal')      ## Matplotlib likes to scale the axes differently, force them to be equal.

##-----
##  Plotting a quiver plot
##-----
ax.quiver(
    x,y,fx,fy,              ## Data being plotted
    pivot='mid',            ## Which part of the arrow is snapped to coordinate
    
    )


plt.show()                  ## plt.show() instead of fig.show() to play nicely with VScode
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8,8)
ax.set_aspect('equal')      ## Matplotlib likes to scale the axes differently, force them to be equal.

##-----
##  Plotting a stream plot
##-----
ax.streamplot(
    x,y,fx,fy,              ## Data being plotted
    density = 2             ## How close the streamlines should be to each other. Default = 1
    )


plt.show()                  ## plt.show() instead of fig.show() to play nicely with VScode
```

## Enhancing with imshow and contour

```python
# We will want a higher resolution
x_range = y_range = np.linspace(0, 2*np.pi, 200)    ## 200 points instead of 10
x,y = np.meshgrid(x_range,y_range)
fx, fy = np.sin(x), np.cos(y)

# We will need a value to represent the magnitude at each coordinate
fmag = np.sqrt(fx**2 + fy**2)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8,8)
ax.set_aspect('equal')      ## Matplotlib likes to scale the axes differently, force them to be equal.


img = ax.imshow(
    fmag,                               ## Data is already in 2D array
    origin='lower',                     ## Where does data start. upper or lower
    extent=(0, 2*np.pi,0, 2*np.pi),     ## Set grid size
    cmap='twilight',                    ## The color map
    
)

# fig.colorbar(img, ax=ax)

ax.streamplot(
    x,y,fx,fy,              ## Data being plotted
    density = 1,            ## How close the streamlines should be to each other. Default = 1
    color='green',
    )


# ax.quiver(
#     ## Rather than plotting 200*200 arrows, only pull every 20th arrow
#     ## Using array slicing (start:stop:step). These are 2D arrays so need to slice twice.
#     x[::20, ::20],
#     y[::20, ::20],
#     fx[::20, ::20],
#     fy[::20, ::20],
#     pivot='mid',
    
#     )
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8,8)
ax.set_aspect('equal')      ## Matplotlib likes to scale the axes differently, force them to be equal.

ax.contour(
    fmag,                               ## Data is already in 2D array
    origin='lower',                     ## Where does data start. upper or lower
    extent=(0, 2*np.pi,0, 2*np.pi),     ## Set grid size
    # cmap='twilight'                     ## The color map
)

# ax.streamplot(
#     x,y,fx,fy,              ## Data being plotted
#     density = 1,            ## How close the streamlines should be to each other. Default = 1
#     # color='green',
#     )

ax.quiver(
    ## Rather than plotting 200*200 arrows, only pull every 20th arrow
    ## Using array slicing (start:stop:step). These are 2D arrays so need to slice twice.
    x[::20, ::20],
    y[::20, ::20],
    fx[::20, ::20],
    fy[::20, ::20],
    pivot='mid',
    
    )
```
