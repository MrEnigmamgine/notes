```python
class si :
    """
    This is a simple class that provides an easy way to access the engineering prefixes for the metric system.
    Example:
        25*si.k results in 25_000
    """
    Y = 10**24  # Yotta
    Z = 10**21  # Zetta
    E = 10**18  # Exa
    P = 10**15  # Peta
    T = 10**12  # Tera
    G = 10**9   # Giga
    M = 10**6   # Mega
    k = 10**3   # kilo

    m = 10**-3  # milli
    µ = 10**-6  # micro
    u = 10**-6  # micro ## so that we can type u instead of µ
    n = 10**-9  # nano
    p = 10**-12 # pico
    f = 10**-15 # femto
    a = 10**-18 # atto
    z = 10**-21 # zepto
    y = 10**-24 # yocto
```

```python
def recipsumrecip(data):
    denom = 0
    for n in data:
        denom += 1/n
    return 1/denom

def recip(num):
    return 1/num
```

```python
s1 = 12
s2 = 18
r1 = 2.7*si.k
r2 = 6.8*si.k
r3 = 6.8*si.k

# Find total current of one source
i_s1 = s1/(series([
    r1,
    parallel([
        r2,
        r3
    ])
]))
# Current divider rule to find current through R2, though could have just divided by 2 since resistances are equal
i_r2_s1 = i_s1*(recip(r2)/sum([recip(r2),recip(r3)]))

# Same thing for other source
i_s2 = s2/(series([
    r3,
    parallel([
        r2,
        r1
    ])
]))
i_r2_s2 = i_s2*(recip(r2)/sum([recip(r2),recip(r1)])) #Ratio of conductance to total conductance

i_r2 = i_r2_s1+i_r2_s2
i_r2
```
